```python
import re
import matplotlib.pyplot as plt

# Path to the pidstat output file
log_file_path = 'cpumem.txt'

# Initialize variables to hold data
timestamps = []
cpu_usages = []
memory_usages = []

# Read the log file
with open(log_file_path, 'r') as file:
    lines = file.readlines()

# Parse the log file
timestamp = 0
for line in lines:
    match = re.search(r'\s+(\d+)\s+(\S+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)\s+(\d+\.\d+)', line)
    if match:
        pid = match.group(1)
        command = match.group(12)
        cpu_usage = float(match.group(9))
        memory_usage = float(match.group(10))
        timestamps.append(timestamp)
        timestamp += 1

# Calculate statistics
if cpu_usages and memory_usages:
    avg_cpu_usage = sum(cpu_usages) / len(cpu_usages)
    max_cpu_usage = max(cpu_usages)
    min_cpu_usage = min(cpu_usages)
    
    avg_memory_usage = sum(memory_usages) / len(memory_usages)
    max_memory_usage = max(memory_usages)
    min_memory_usage = min(memory_usages)
else:
    avg_cpu_usage = max_cpu_usage = min_cpu_usage = 0
    avg_memory_usage = max_memory_usage = min_memory_usage = 0

# Print the results
print(f'Average CPU Usage: {avg_cpu_usage:.2f}%')
print(f'Max CPU Usage: {max_cpu_usage:.2f}%')
print(f'Min CPU Usage: {min_cpu_usage:.2f}%')
print(f'Average Memory Usage: {avg_memory_usage:.2f} KB')
print(f'Max Memory Usage: {max_memory_usage:.2f} KB')
print(f'Min Memory Usage: {min_memory_usage:.2f} KB')

# Plot CPU and memory usage
fig, ax1 = plt.subplots()

ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('CPU Usage (%)', color='tab:blue')
ax1.plot(timestamps, cpu_usages, color='tab:blue', marker='o', linestyle='-', label='CPU Usage')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Memory Usage (KB)', color='tab:red')
ax2.plot(timestamps, memory_usages, color='tab:red', marker='x', linestyle='-', label='Memory Usage')
ax2.tick_params(axis='y', labelcolor='tab:red')

fig.tight_layout()
plt.title('CPU and Memory Usage of zeekctl Over Time')
plt.grid(True)
plt.show()
```
